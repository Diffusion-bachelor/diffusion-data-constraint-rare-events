```python
from collections import Counter

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

dataset = [
    {"text": "Hello world."},
    {"text": "Hello there!"},
    {"text": "The world is big."},
    {"text": "World was there!! Crazy escaping."},
    {"text": "Hello world again."}
]

token_counts = Counter()

for example in dataset:
    text = example["text"]

    # tokenize text
    tokens = tokenizer.tokenize(text)   # returns token strings directly

    for token in tokens:
        token = token.lstrip("Ġ").strip()
        token_counts[token] += 1

for token, count in token_counts.most_common(20):
    print(f'"{token.lstrip("Ġ")}": {count}')
```
