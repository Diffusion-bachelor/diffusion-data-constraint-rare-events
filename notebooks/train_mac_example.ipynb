# Tiny Mac Training Example

This notebook shows a small end-to-end example for training a tiny language model with `train_mac.py` on macOS (MPS) or CPU.

The goal is quick verification, not full-scale reproduction.

```python
# Basic imports used throughout this notebook
from pathlib import Path
import subprocess
import sys
import os
```

## Locate Project Root

This cell finds the workspace root so the notebook can call the training script reliably.

```python
# Assume the notebook is in /workspace/notebooks
cwd = Path.cwd().resolve()
repo_root = cwd if (cwd / 'train_mac.py').exists() else cwd.parent

train_script = repo_root / 'train_mac.py'
data_file = repo_root / 'data' / 'c4_sample.jsonl'

print('Working dir :', cwd)
print('Repo root   :', repo_root)
print('Script exists:', train_script.exists())
print('Data exists  :', data_file.exists())
```

## Example Configuration

We keep the model and dataset slice small so it can run quickly on a laptop.

```python
# Tiny configuration for a fast demonstration run
example_args = [
    sys.executable,
    str(train_script),
    '--epochs', '3',
    '--max-samples', '100',
    '--n-embd', '64',
    '--n-layer', '2',
    '--n-head', '4',
    '--batch-size', '16',
    '--block-size', '32',
]

print('Command:')
print(' '.join(example_args))
```

## Run Training

This executes one tiny training job and prints the final lines (including validation perplexity and checkpoint path).

```python
# Run the command and capture output
proc = subprocess.run(
    example_args,
    cwd=str(repo_root),
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    text=True,
    check=False,
)

print('Exit code:', proc.returncode)
print('--- Last 40 lines of output ---')
print('\n'.join(proc.stdout.splitlines()[-40:]))
```
